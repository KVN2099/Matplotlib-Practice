## Import Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1, 2, 3]
y = [2, 4, 6]

font = {'fontname': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

# Line 1
plt.figure(figsize=(16, 6), dpi=144)
plt.plot(x, y, label='2x', linewidth=3, linestyle='-.', marker='^', markersize=14, markeredgecolor='red', color='gray')
plt.xticks([0.01, 2, 5])
plt.yticks((1, 3, 5, 7))
plt.title('My first graph', fontdict=font)
plt.xlabel('X axis', fontdict=font)
plt.ylabel('Y axis', fontdict=font)

# Line 2
plt.figure(figsize=(16, 6), dpi=144)
x2 = np.arange(0, 4, 0.1)
plt.plot(x2[:12], x2[:12]**2, 'g*-', label='x^2 1')
plt.plot(x2[12:], x2[11:]**2, marker='o', label='x^2 2', markersize=5)

plt.xticks(np.arange(0, 4, 0.5))
plt.yticks(np.arange(0, 16, 1))

plt.grid()
plt.legend()
plt.show()
```
